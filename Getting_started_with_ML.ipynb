```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

```

```python
"""
define layers using Sequential
only 1 line so there is only 1 neuron

keras.layers
Our Dense layer has units=1 specified, so we have just one dense
layer with one neuron in our entire neural network.

"""
model = Sequential([Dense(units=1, input_shape=[1])])

"""
We can use any optimizer but we picked stochastic gradient descent
when given the values, the previous guess, and the results of calculating the errors (or loss) on that guess, it can generate another one
"""
model.compile(optimizer='sgd', loss='mean_squared_error')
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
model.fit(xs, ys, epochs=500) # epochs is amount of passes
print(model.predict([10.0])) #predict y when x is 10
```
